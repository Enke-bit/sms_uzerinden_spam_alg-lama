```python
# Kelime frekansı analizi için kelime sayısını hesaplama
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['message'])
word_freq = X.sum(axis=0)
words = vectorizer.get_feature_names_out()
word_freq_df = pd.DataFrame(word_freq.T, index=words, columns=['count']).sort_values(by='count', ascending=False)

plt.figure(figsize=(14, 10))
top_words = word_freq_df.head(20)
sns.barplot(x=top_words['count'], y=top_words.index, palette='viridis')
plt.title('Top 20 Most Frequent Words in Messages')
plt.xlabel('Count')
plt.ylabel('Word')
plt.show()
```

```python
# Veri ön işleme
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
# Mesajları ve etiketleri ayırma
X = df['message']
y = df['label']
```

```python
# Metin verilerini sayısal verilere dönüştürme (Bag of Words)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)
```

```python
# Eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Modeli eğitme
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train, y_train)
```

```python
# Modeli test et ve tahminlerde bulun
predictions = naive_bayes.predict(X_test)
```

```python
# Sonuçları yazdırma
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)
```

```python
print(f"Model Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)
```
